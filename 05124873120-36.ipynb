a. Buatlah deskripsi statistik visualisasi BoxPlot untuk menjelaskan Top 5 Product 
dengan penjualan terbanyak. Jelaskan grafik tersebut

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

path_cust='https://raw.githubusercontent.com/hackathonBI/CCS/master/sample%20data/customers.csv' 
dfcustomer = pd.read_csv(path_cust) 

path_gas='https://raw.githubusercontent.com/hackathonBI/CCS/master/sample%20data/gasstations.csv'
dfgasstation = pd.read_csv(path_gas)

path_product='https://raw.githubusercontent.com/hackathonBI/CCS/master/sample%20data/products.csv'
dfproducts = pd.read_csv(path_product) 

path_sample='https://raw.githubusercontent.com/hackathonBI/CCS/master/sample%20data/sample.csv'
dfsamples = pd.read_csv(path_sample) 

#group product by price
tmp3 = dfsamples.groupby(['ProductID'],as_index=False).sum().sort_values('Price',ascending=False)
tmp3b = pd.merge(tmp3,dfproducts, left_on='ProductID', right_on='ProductID')

#get top
top = tmp3b.rename(columns={"Price":"Sales"})
top5 = top[["Description","Amount","Sales","ProductID"]].set_index("ProductID").head(5)
display(top5)

#merge
all = dfsamples.merge(dfproducts, left_on='ProductID', right_on='ProductID')
all = all.merge(dfgasstation, left_on='GasStationID', right_on='SiteID')
all = all.merge(dfcustomer, left_on='CustomerID', right_on='CustomerID')
display(all)

#getAll product
allByProduct = all.copy()
allByProduct = allByProduct.set_index('ProductID')
top5Only = allByProduct[allByProduct.index.isin(top5.index)]
display(allByProduct)

#plot
sns.boxplot(x="Description", y="Amount", data=top5Only)
```

c. Buatlah line chart yang didalamnya terdapat variable jumlah sales setiap jam. 
Vairable jam atau waktu pada sumbu x dan variable jumlah sales pada jam tersebut 
pada sumbu y.

Variable jam dapat anda peroleh dari melakukan ekstraksi variable waktu pada 
kolom time. Jika terdapat data yang hilang atau tidak ada, anda bisa menggunakan 
Teknik re-sampling pada modul 8 untuk melengkapi data tersebut

```python
# Resampling
all['DateTime'] = pd.to_datetime(all['Date'] + ' ' +  all['Time'], format='%Y-%m-%d %H:%M:%S')
#display(all)
allByHour = all[["DateTime","Amount"]].resample('60min', on='DateTime').sum()

#Group by Hours
allByHour = allByHour.reset_index()
allByHour['Hour'] = allByHour['DateTime'].dt.hour
allByHour = allByHour[['Hour','Amount']]

#Plot Line Chart
plt.plot( allByHour['Hour'], allByHour['Amount'], marker='o', label='Amount' )
plt.ylabel('Amount') 
plt.xlabel('Hour') 
plt.legend()
plt.show()
```
