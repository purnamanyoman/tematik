**DATA UNDERSTANDING : PREVIEW DATASET**

```python
#Baca data dari customers.csv
df = pd.concat(map(pd.read_csv, glob.glob(os.path.join('', "customers.csv"))))
display(df)
```

```python
#Baca data dari gasstations.csv
df2 = pd.concat(map(pd.read_csv, glob.glob(os.path.join('', "gasstations.csv"))))
print(df2)
```

```python
#Baca data dari products.csv
df3 = pd.concat(map(pd.read_csv, glob.glob(os.path.join('', "products.csv"))))
print(df3)
```

```python
#Baca data dari trancations_1k.csv
df4 = pd.concat(map(pd.read_csv, glob.glob(os.path.join('', "transactions_1k.csv"))))
print(df4)
```

**DATA UNDERSTANDING : MERGE TABLES**

```python
merge1 = pd.merge(df4, df)
merge2 = pd.merge(merge1,df2, left_on='GasStationID', right_on='GasStationID')
merge3 = pd.merge(merge2,df3, left_on='ProductID', right_on='ProductID')
display(merge3)
```

**DATA UNDERSTANDING : PREVIEW DATA TYPES**

```python
merge3.dtypes
```

**Jawaban Nomor 1** Top 5 Customer Yang Memiliki Nilai Transaksi Paling Banyak

```python
top5 = merge3.head(5)
display(top5.set_index("CustomerID"))
```

```python
tmp1 = merge3.groupby(['CustomerID'],as_index=False).sum().sort_values('Price',ascending=False)
tmp1 = tmp1.rename(columns={"Price":"Gas Consumption"})

display(tmp1[["CustomerID","Amount","Gas Consumption"]].set_index("CustomerID").head(5))
```

**Jawaban Nomor 2** : Top 5 Gas Stations yang memiliki nilai transaksi paling banyak (Price)

```python
tmp2 = merge3.groupby(['GasStationID'],as_index=False).sum().sort_values('Price',ascending=False)
tmp2 = tmp2.rename(columns={"Price":"Gas Sales"})

display(tmp2[["GasStationID","Amount","Gas Sales"]].set_index("GasStationID").head(5))
```

**Jawaban Nomor 3** : Top 5 Jenis Produk Yang Memiliki Nilai Transaksi paling Banyak (Amount)

```python
tmp3 = merge3.groupby(['ProductID'],as_index=False).sum().sort_values('Price',ascending=False)
tmp3b = pd.merge(tmp3,df3, left_on='ProductID', right_on='ProductID')
tmp3b = tmp3b.rename(columns={"Price":"Sales"})
display(tmp3b[["Description","Amount","Sales"]].set_index("Description").head(5))
```

**Jawaban Nomor 4** : Deskripsi Statistik Untuk Masing-Masing Hari (23,24,25,26) Pada Dataset transaction_1k

```python
tmp4 = merge3.rename(columns={"Price":"Sales"})
tmp4 = tmp4.groupby(['Date'],as_index=True)
# tmp4b = tmp4.rename(columns={"Price":"Sales"})
# tmp4b = tmp4b[['Date','Amount','Price']].set_index('Date')

display(tmp4[['Date','Amount','Sales']].describe())
# tmp4b.describe()
```

**Jawaban Nomor 5** : Waktu Terbaik (Hari dan Jam) Dimana Paling Banyak User Gas Station Melakukan Transaksi

```python
#Dibuat untuk mencari jam dengan jumlah customer pada hari Tertentu 
tmp5 = merge3.rename(columns={"Price":"Gas Sales"})

tmp5['Time'] = pd.to_datetime(tmp5['Time'], format='%H:%M:%S').dt.hour
tmp5b = tmp5.groupby(["Date","Time"])['Amount','Gas Sales'].sum()
tmp5c = tmp5b[['Amount','Gas Sales']]
display(tmp5c)

#Dibuat untuk mencari jam dengan customer terbanyak pada hari Tertentu 
tmp5d = tmp5c.sort_values(['Date','Amount'],ascending=[True,False]).reset_index()
display(tmp5d.drop_duplicates(subset="Date").set_index(['Date']))
```
