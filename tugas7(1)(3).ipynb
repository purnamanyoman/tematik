# Latihan 2: Permen 
Kita  memiliki  sekantong  permen.  Terdapat  lima  jenis  permen,  masing-masing  diberi 
nama di bawah ini. Buat diagram yang menunjukkan persentase peluang bahwa kita akan 
mengeluarkan  permen  Snickers  dari  kantong  jika  kita  melakukan  pengambilan  acak. 
Sebutkan peluang memilih permen Snickers

```python
import matplotlib.pyplot as plt

candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix']
candy_counts = [52, 39, 78, 13, 78]
explode = (0, 0.15, 0, 0, 0)
plt.figure(figsize=(14,7))
plt.pie(
    candy_counts,
    labels= candy_names,
    autopct='%1.1f%%',
    explode=explode,
    shadow='True',
    startangle=90
)
plt.title('Peluang Snickers diambil 1x')
plt.show()
```

Untuk menampilkan pesentase setiap kategorinya, saya menggunakan pie chart, karena lebih mudah untuk mengamati dan menganalisanya.Berapa persen kemungkinan Anda akan memilih Snickers saat mengeluarkan permen dari tas secara acak yakni Sebanyak 15%

```python
dessert_sales = { 
  'Lava Cake': 14, 
  'Mousse': 5, 
  'Chocolate Cake': 12, 
    'Ice Cream': 19, 
  'Truffles': 6, 
  'Brownie': 8, 
  'Chocolate Chip Cookie': 12, 
  'Chocolate Pudding': 9, 
  'Souffle': 10, 
  'Chocolate Cheesecake': 17, 
  'Chocolate Chips': 2, 
  'Fudge': 9, 
  'Mochi': 13, 
}
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df= pd.DataFrame([dessert_sales.keys(), dessert_sales.values()]).T
df.columns= ['Makanan', 'Terjual'] 
```

```python
df
```
